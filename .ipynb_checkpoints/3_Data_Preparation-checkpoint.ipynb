```python
# drop duplicated index: 'Unnamed: 0'
df2 = df1.drop(['Unnamed: 0', 'sensor_15', 'timestamp'] , axis='columns')

# replace machine status values in dataframe column: Normal = 0, Recovering = 0, Broken = 1
df2.machine_status.replace({'NORMAL': '0', 'RECOVERING': '0', 'BROKEN': '1'}, inplace=True)
```

<div class="alert alert-warning">
<H3>Findings </H3>

Steps involved:
    <ol>
        <li>Dropped 'sensor_15': This field was dropped as it was a misreading and could skew data if or once tunred back on. Being that any model would have to be re-trained anyway it does not make sense now to add unneccesary noise.</li>
        <li>Replaced machine_status target column's values from alpha to numeric: Normal = 0, Recovering = 1, Broken = 1</li>
        <li>Dropped 'timestamp': As we are only concerned with predicting machine failure based on sensor readings in an unbalanced data set the time stamp will add no value to analysis given the scope of this exercise: time series data will most likely be broken as the undersampling process will remove disparate chunks of time</li>
        </ol>

```python
# change target variable data type from object to integer
df2['machine_status'] = df2['machine_status'].astype(float)

# fill nan values with column means
df3 = df2.iloc[:,0:50].fillna(df2.mean())

# concatenate target_values back to cleaned dependent variable data set
df4 = pd.concat([df3, df2.machine_status], axis=1)
```

```python
# MOVE TO MODELING
# pull out the target variable
target = 'machine_status'

# create x (independent) and y(target) variables
X = df4.drop(target, axis=1) 
Y = df4[target].values

# split data into test and training sets
# set seed to 42 so no matter how many times you execute your code the result would be the same . i.e, same values in train and test datasets.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
                                                    test_size=0.33,
                                                    random_state=42)

# visualize the balance of target variable
ax = sns.countplot(x=target, data=df4)
print(df4[target].value_counts())
```

```python
# calculate percentage of class inbalances on target variable (1 = FAILURE)
100 * (7/float(df4.shape[0]))
```

```python
from sklearn.utils import resample

# Separate majority and minority classes
df_majority = df4[df4[target]==0]
df_minority = df4[df4[target]==1]
 
# Upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=73438, # to match majority class (1/3 ratio)
                                 random_state=123) # reproducible results
 
# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
 
# Display new class counts
df_upsampled[target].value_counts()
```

```python
# visualize the balance of target variable
ax = sns.countplot(x=target, data=df_upsampled)
print(df_upsampled[target].value_counts())
```

Steps involved:
    <ol>
        <li>In order to build a logistic regression model it was important to ensure that the x-variables are of a consistent shape with the y-variables. Additionally, it was critical that management normalized the understanding of the y-variable: 'NORMAL': '0', 'RECOVERING': '0', 'BROKEN': '1', so that the target being predicted is consistent.</li>
        <li>Filled nan values with column means</li>
        <li>Up-sampled normal readings 1:4 to reduce the amplification of unbalanced data</li>
        </ol>
